```python
from peewee import *
db = MySQLDatabase(
    "todo" , 
    host = 'localhost',
    user='root',
    password='123123',
)
db.connect()
print('Connected')
   
```

```python
import datetime
class tasks(Model):
    id = AutoField()
    task_name = CharField()
    status = CharField()
    datatime = DateTimeField(default=datetime.datetime.now)
    class Meta:
        database = db

db.create_tables([tasks])

        
```

```python
!pip install django-datetime-widget
```

```python
data = [
    ['Tiem work','Done'],
     ['Swime','InProgres'],
     ['Tiem Codeing', 'Done'],
]
for x in data:
    tasks.create(task_name=x[0],status=x[1])
```

```python
data = tasks.select()
for x in data:
    print(x.id,x.task_name,x.status,x.datatime)
```
